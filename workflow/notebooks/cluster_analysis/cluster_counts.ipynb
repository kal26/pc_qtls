```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import os
import ast
import re

import upsetplot as up
from tqdm.auto import tqdm  # for notebooks

# Create new `pandas` methods which use `tqdm` progress
# (can use tqdm_gui, optional kwargs, etc.)
tqdm.pandas()

# set higher figure resolution (for poster)
import matplotlib as mpl
#mpl.rcParams['figure.dpi']= 1000
```

```python
# get outputs from a config file
prefix = '/home/klawren/oak/pcqtls'
import yaml
config_path= f'{prefix}/config/new_clusters/proteincoding_rewrite.yaml'
with open(config_path, 'r') as f:
    config = yaml.safe_load(f)

tissue_id_path = config['tissue_id_path']
clusters_dir = config['clusters_dir']
expression_dir = config['expression_dir']

# load in the tissue ids 
tissue_df = pd.read_csv(f"{prefix}/{tissue_id_path}", header=0)
tissue_ids = list(tissue_df['Tissue'])
```

### load in data

```python
# load in the clusters data
clusters = {}
for tissue_id in tissue_df['Tissue']:
    clusters[tissue_id] = pd.read_csv(f'{prefix}/{clusters_dir}/{tissue_id}_clusters_all_chr.csv')
clusters_df = pd.concat(clusters, keys=tissue_df['Tissue'], names=('tissue_id', 'idx')).reset_index('idx', drop=True)
```

```python
# load in expression data (so we only compare to pairs of genes also expressed in this tissue)
# load in expression data for all tissues 
expressions={}
for tissue_id in tissue_ids: 
    expressions[tissue_id] = pd.read_csv(f'{prefix}/{expression_dir}/{tissue_id}.v8.normalized_expression.bed', sep='\t')
    expressions[tissue_id] = expressions[tissue_id].sort_values(['#chr', 'start', 'end'])
    # we only need the transcript ids from the expression dir 
    expressions[tissue_id].drop(columns=expressions[tissue_id].columns[4:], inplace=True)
    
expression_df = pd.concat(expressions, keys=tissue_df['Tissue'], names=('tissue_id', 'idx')).reset_index('idx', drop=True)
```

```python
tissue_pal = sns.color_palette('blend:#B83A4B,#FEC51D,#67AFD2', n_colors=len(tissue_ids))
tissue_pal
```

### What fraction of genes are in clusters?

## Clusters per tissue

```python
# create a dataframe with the number of clusters for each tissue 
cs_counts_df = pd.DataFrame({'tissue':tissue_df['Tissue'], 
                       'num_clusters':[sum(clusters_df['Tissue'] == t) for t in tissue_df['Tissue']]})
```

```python
# plot showing number of clusters in each tissue

fig, ax = plt.subplots(figsize=(3,5))

ax = sns.barplot(cs_counts_df, x='num_clusters', hue='tissue', y='tissue', palette=tissue_pal, saturation=1, edgecolor='k', width=.85, ax=ax)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Replace underscores with spaces in the tissue labels
tissue_labels = ax.get_yticklabels()
tissue_labels = [tissue_label.get_text().replace('_', ' ') for tissue_label in tissue_labels]
ax.set_yticklabels(tissue_labels)

ax.set_xlabel('Number of clusters')
ax.set_ylabel('')

plt.show()
```

## Fraction of genes in clusters

```python
frac_genes_in_clusters=[]
for tissue_id in tissue_ids: 
    all_cluster_transcripts = np.concatenate(clusters[tissue_id]['Transcripts'].str.split(',').values)
    frac_genes_in_clusters.append(sum(expressions[tissue_id]['gene_id'].isin(all_cluster_transcripts))/len(expressions[tissue_id]))

frac_df = pd.DataFrame({'tissue_id': tissue_ids, 'fraction':frac_genes_in_clusters})
print(frac_df['fraction'].mean())

# plot
sns.barplot(pd.DataFrame({'tissue_id': tissue_ids, 'fraction':frac_genes_in_clusters}), x='fraction', y='tissue_id', palette=tissue_pal, saturation=1, edgecolor='k', width=.85)
plt.xlabel('Fraction expressed genes in clusters')
plt.ylabel('')
plt.show()
```

## Size distribution of clusters

```python
# plot showing distribution of clsuter sizes in each tissue
clusters_df['num_genes'] = clusters_df['Transcripts'].str.split(',').apply(len)


fig, ax = plt.subplots(figsize=(6,5))


PROPS = {
    'boxprops':{ 'edgecolor':'k'},
    'medianprops':{'color':'k'},
    'whiskerprops':{'color':'k'},
    'capprops':{'color':'k'}, 
    'flierprops':{'markerfacecolor':'k', 'markersize':2,'linestyle':'none', 'markeredgecolor':'k',}}

sns.boxplot(clusters_df[clusters_df['num_genes']>2], x='num_genes', hue='Tissue', y='Tissue', ax=ax, palette=tissue_pal, saturation=1, width=.85, **PROPS)

# remove labels
#ax.set_yticklabels('')

ax.set_xlabel('Number of genes in cluster')
#ax.set_ylabel('')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

## Number of tissues each cluster gene is in a cluster in

```python
# expressed gene sharing across tissues
expression_memberships_df = expression_df.reset_index().groupby('gene_id').agg({'tissue_id':list})
expression_upset = up.from_memberships(expression_memberships_df['tissue_id'])

# plot
up.plot(expression_upset, subset_size='count', min_subset_size=15, show_counts=True, sort_by='-degree', sort_categories_by='-input')
plt.title('Sharing of expressed genes across tissues')
```

```python
# counts for number of tissues each gene is expressed in 

# get the counts 
expression_tissue_counts = expression_df.reset_index().groupby('gene_id').agg({'tissue_id':pd.Series.nunique})

# plot
sns.countplot(expression_tissue_counts, x='tissue_id')
plt.xlabel('Number tissues')
plt.ylabel('Number genes')
plt.title('Number of tissues genes are expressed in')
```

```python
# sharing of cluster genes across tissues

# mebership df for genes in clusters
clusters_df['transcipt_id'] = clusters_df['Transcripts'].str.split(',')
cluster_gene_memberships_df = clusters_df.explode('transcipt_id').reset_index('tissue_id').groupby('transcipt_id').agg({'tissue_id':list})

# boolean label version 
cluster_gene_upset = up.from_memberships(cluster_gene_memberships_df['tissue_id'])

# plot
up.plot(cluster_gene_upset, subset_size='count', min_subset_size=15, show_counts=True, sort_by='-degree', sort_categories_by='-input')
plt.title('Sharing of cluster genes across tissues')
```

```python
# counts for number of tissues each gene is in a cluster in (does'nt have to be the same cluster)

# get the counts 
cluster_gene_tissue_counts = clusters_df.explode('transcipt_id').reset_index('tissue_id').groupby('transcipt_id').agg({'tissue_id':pd.Series.nunique})

# color by whether a gene is expressed in all tissues or not
expression_tissue_counts['expressed_in'] = 'some'
expression_tissue_counts.loc[expression_tissue_counts['tissue_id'] == 13,'expressed_in'] = 'all'
expression_tissue_counts.loc[expression_tissue_counts['tissue_id'] == 1,'expressed_in'] = 'one'

cluster_gene_tissue_counts = cluster_gene_tissue_counts.merge(expression_tissue_counts['expressed_in'], right_index=True, left_index=True)

# plot
sns.countplot(cluster_gene_tissue_counts, x='tissue_id', hue='expressed_in')
plt.xlabel('Number tissues')
plt.ylabel('Number genes')
plt.title('Count of genes vs number tissues the gene is in a cluster in')
```

```python
# sharing of the exact clusters
cluster_memberships_df = clusters_df.reset_index('tissue_id').groupby('Transcripts').agg({'tissue_id':list})

# boolean label version 
cluster_upset = up.from_memberships(cluster_memberships_df['tissue_id'])

# plot
up.plot(cluster_upset, subset_size='count', min_subset_size=15, show_counts=True, sort_by='-degree', sort_categories_by='-input')
plt.title('Sharing of exact clusters across tissues')
```

```python
# counts for number of tissues each gene is in a cluster in (does'nt have to be the same cluster)

# get the counts 
cluster_tissue_counts = clusters_df.reset_index('tissue_id').groupby('Transcripts').agg({'tissue_id':pd.Series.nunique})

# plot
sns.countplot(cluster_tissue_counts, x='tissue_id')
plt.xlabel('Number tissues')
plt.ylabel('Number clusters')
plt.title('Number of tissues exact same clusters are in')
```
