```python

def get_counts(config):
    counts = pd.DataFrame()
    for tissue_id in tissue_ids:
        cluster_df = load_cluster(config, tissue_id)
        counts.loc[tissue_id, 'clusters'] = len(cluster_df)
        counts.loc[tissue_id, 'cluster_genes'] = sum(cluster_df['N_genes'])

        # load eqtls and analyze
        try: 
            e_permutation_df = load_e_cis(config, tissue_id)
            e_permutation_signifigant = e_permutation_df[e_permutation_df['qval'] < .05]
            counts.loc[tissue_id, 'e_clusters_permutation'] = e_permutation_signifigant['cluster_id'].nunique()
            counts.loc[tissue_id, 'e_qtl_permutation'] = e_permutation_signifigant['variant_id'].nunique()
        except FileNotFoundError as e:
            print(e)
            counts.loc[tissue_id, 'e_clusters_permutation'] = 0
            counts.loc[tissue_id, 'e_qtl_permutation'] = 0

        # load pcqlts and anaylze
        try: 
            pc_permutation_df = load_pc_cis(config, tissue_id)
            pc_permutation_signifigant = pc_permutation_df[pc_permutation_df['qval'] < .05]
            counts.loc[tissue_id, 'pc_clusters_permutation'] = pc_permutation_signifigant['cluster_id'].nunique()
            counts.loc[tissue_id, 'pc_qtl_permutation'] = pc_permutation_signifigant['variant_id'].nunique()


        except FileNotFoundError as e:
            print(e)
            counts.loc[tissue_id, 'pc_clusters_permutation'] = 0
            counts.loc[tissue_id, 'pc_qtl_permutation'] = 0


    return counts.rename_axis('Tissue').reset_index()
```

```python

def get_pc1_only_counts(config, cluster_cutoff=2):
    counts = pd.DataFrame()
    for tissue_id in tissue_ids:
        cluster_df = load_cluster(config, tissue_id)
        cluster_df = cluster_df[cluster_df['cluster_id'].str.split('_').apply(len) > cluster_cutoff]
        counts.loc[tissue_id, 'clusters'] = len(cluster_df)
        counts.loc[tissue_id, 'cluster_genes'] = sum(cluster_df['N_genes'])

        # load eqtls and analyze
        e_permutation_df = load_e_permutation(config, tissue_id)
        e_permutation_signifigant = e_permutation_df[e_permutation_df['pval_perm'] < .05]
        e_permutation_signifigant = e_permutation_signifigant[e_permutation_signifigant['cluster_id'].str.split('_').apply(len) > cluster_cutoff]
        counts.loc[tissue_id, 'e_clusters_permutation'] = e_permutation_signifigant['cluster_id'].nunique()
        counts.loc[tissue_id, 'e_qtl_permutation'] = e_permutation_signifigant['variant_id'].nunique()

        # load pcqlts and anaylze
        pc_permutation_df = load_pc_permutation(config, tissue_id)
        pc_permutation_signifigant = pc_permutation_df[pc_permutation_df['pval_perm'] < .05]
        pc_permutation_signifigant = pc_permutation_signifigant[pc_permutation_signifigant['cluster_id'].str.split('_').apply(len) > cluster_cutoff]
        counts.loc[tissue_id, 'pc_clusters_permutation'] = pc_permutation_signifigant['cluster_id'].nunique()
        counts.loc[tissue_id, 'pc_qtl_permutation'] = pc_permutation_signifigant['variant_id'].nunique()

        # load pcqlts and anaylze
        pc1_permutation_df = load_pc_permutation(config, tissue_id, pc1_only=True)
        pc1_permutation_signifigant = pc1_permutation_df[pc1_permutation_df['pval_perm'] < .05]
        pc1_permutation_signifigant = pc1_permutation_signifigant[pc1_permutation_signifigant['cluster_id'].str.split('_').apply(len) > cluster_cutoff]
        counts.loc[tissue_id, 'pc1_clusters_permutation'] = pc1_permutation_signifigant['cluster_id'].nunique()
        counts.loc[tissue_id, 'pc1_qtl_permutation'] = pc1_permutation_signifigant['variant_id'].nunique()


    return counts.rename_axis('Tissue').reset_index()
```

## high corr clusters only?

I can count the number of signifigant hits by filtering the cis_mapping output and recauclating z scores 

```python
def get_high_corr_clusters(config, tissue_id, corr_cutoff = .5):
    clusters = load_cluster(config, tissue_id)
    clusters['abs_cor'] = np.where(clusters['Mean_neg_cor'].isna(), clusters['Mean_pos_cor'], - clusters['Mean_neg_cor'])
    clusters['abs_cor'] = np.where(clusters['abs_cor'] < clusters['Mean_pos_cor'], clusters['Mean_pos_cor'], clusters['abs_cor'])
    high_cor_clusters = clusters[clusters['abs_cor'] > corr_cutoff]
    return high_cor_clusters
```

```python
def get_high_corr_counts(config, corr_cutoff = .5):

    counts = pd.DataFrame()
    for tissue_id in tissue_ids:
        high_corr_clusters = get_high_corr_clusters(config, tissue_id, corr_cutoff = corr_cutoff)
        counts.loc[tissue_id, 'clusters'] = len(high_corr_clusters)
        counts.loc[tissue_id, 'cluster_genes'] = sum(high_corr_clusters['N_genes'])

        # get e for high corr only
        e_cis_df = load_e_cis(config, tissue_id)
        high_corr_e = e_cis_df[e_cis_df['cluster_id'].isin(high_corr_clusters['cluster_id'])]
        tensorqtl.calculate_qvalues(high_corr_e, qvalue_lambda=0.85)
        counts.loc[tissue_id, 'cluster_e_sig'] = len(high_corr_e[high_corr_e['qval'] < .05]['cluster_id'].unique())
        counts.loc[tissue_id, 'qtl_e_sig'] = len(high_corr_e[high_corr_e['qval'] < .05]['variant_id'].unique())

        pc_cis_df = load_pc_cis(config, tissue_id)
        high_corr_pc = pc_cis_df[pc_cis_df['cluster_id'].isin(high_corr_clusters['cluster_id'])]
        tensorqtl.calculate_qvalues(high_corr_pc, qvalue_lambda=0.85)
        counts.loc[tissue_id, 'cluster_pc_sig'] = len(high_corr_pc[high_corr_pc['qval'] < .05]['cluster_id'].unique())
        counts.loc[tissue_id, 'qtl_pc_sig'] = len(high_corr_pc[high_corr_pc['qval'] < .05]['variant_id'].unique())


        high_corr_pc1 = high_corr_pc[high_corr_pc.index.str.contains('pc1')]
        tensorqtl.calculate_qvalues(high_corr_pc1, qvalue_lambda=0.85)
        counts.loc[tissue_id, 'cluster_pc1_sig'] = len(high_corr_pc1[high_corr_pc1['qval'] < .05]['cluster_id'].unique())
        counts.loc[tissue_id, 'qtl_pc1_sig'] = len(high_corr_pc1[high_corr_pc1['qval'] < .05]['variant_id'].unique())

    return counts.rename_axis('Tissue').reset_index()
```

```python
def get_high_corr_count_df(corr_cutoff, subsample_nums):
    sub_high_corr_counts = []

    for num_samples in tqdm(subsample_nums):
        sub_high_corr_counts.append(get_high_corr_counts(sub_config[num_samples], corr_cutoff))

    # normalize by number cluster-genes
    high_corr_counts = pd.concat(sub_high_corr_counts, keys=subsample_nums, names=['num_samples', 'idx'])
    
    high_corr_counts['e_clusters_permutation_normed'] = high_corr_counts['cluster_e_sig']/high_corr_counts['clusters']
    high_corr_counts['pc_clusters_permutation_normed'] = high_corr_counts['cluster_pc_sig']/high_corr_counts['clusters']
    high_corr_counts['pc1_clusters_permutation_normed'] = high_corr_counts['cluster_pc1_sig']/high_corr_counts['clusters']

    high_corr_counts.reset_index(level=0, inplace=True)
    return high_corr_counts
```

```python
def get_sankey_part(all_clusters, subsetsize1, subsetsize2, total_clusters):
    part_df = up.query(up.from_contents({f'{subsetsize1}':all_clusters[subsetsize1], f'{subsetsize2}':all_clusters[subsetsize2]})).subset_sizes
    part_df.loc[False, False] = total_clusters - part_df.sum()
    part_df = pd.DataFrame(part_df).reset_index()
    part_df.loc[part_df[f'{subsetsize2}'] == True, f'{subsetsize2}'] = f'QTL w/ {subsetsize2}'
    part_df.loc[part_df[f'{subsetsize2}'] == False, f'{subsetsize2}'] = f'no QTL w/ {subsetsize2}'
    part_df.loc[part_df[f'{subsetsize1}'] == True, f'{subsetsize1}'] = f'QTL w/ {subsetsize1}'
    part_df.loc[part_df[f'{subsetsize1}'] == False, f'{subsetsize1}'] = f'no QTL w/ {subsetsize1}'
    part_df = part_df.rename(columns={f'{subsetsize1}':'Source', f'{subsetsize2}':'Target', 'size':'Value'})
    return part_df
```

```python
def get_sankey_part_pce(e_clusters, pc_clusters, subsetsize1, subsetsize2, total_clusters):
    part_df = up.query(up.from_contents({f'{subsetsize1} eqtl':e_clusters[subsetsize1],
                                        f'{subsetsize1} pcqtl':pc_clusters[subsetsize1],
                                        f'{subsetsize2} eqtl':e_clusters[subsetsize2],
                                        f'{subsetsize2} pcqtl':pc_clusters[subsetsize2]})).subset_sizes
    part_df.loc[False, False, False, False] = total_clusters - part_df.sum()
    part_df = pd.DataFrame(part_df).reset_index()

    part_df.loc[(part_df[f'{subsetsize2} eqtl']==True) & (part_df[f'{subsetsize2} pcqtl']==True), 'Target'] = f'eQTL and pcQTL w/ {subsetsize2}'
    part_df.loc[(part_df[f'{subsetsize2} eqtl']==True) & (part_df[f'{subsetsize2} pcqtl']==False), 'Target'] = f'eQTL w/ {subsetsize2}'
    part_df.loc[(part_df[f'{subsetsize2} eqtl']==False) & (part_df[f'{subsetsize2} pcqtl']==True), 'Target'] = f'pcQTL w/ {subsetsize2}'
    part_df.loc[(part_df[f'{subsetsize2} eqtl']==False) & (part_df[f'{subsetsize2} pcqtl']==False), 'Target'] = f'no QTL w/ {subsetsize2}'

    part_df.loc[(part_df[f'{subsetsize1} eqtl']==True) & (part_df[f'{subsetsize1} pcqtl']==True), 'Source'] = f'eQTL and pcQTL w/ {subsetsize1}'
    part_df.loc[(part_df[f'{subsetsize1} eqtl']==True) & (part_df[f'{subsetsize1} pcqtl']==False), 'Source'] = f'eQTL w/ {subsetsize1}'
    part_df.loc[(part_df[f'{subsetsize1} eqtl']==False) & (part_df[f'{subsetsize1} pcqtl']==True), 'Source'] = f'pcQTL w/ {subsetsize1}'
    part_df.loc[(part_df[f'{subsetsize1} eqtl']==False) & (part_df[f'{subsetsize1} pcqtl']==False), 'Source'] = f'no QTL w/ {subsetsize1}'

    part_df = part_df.rename(columns={'size':'Value'})
    return part_df[['Source', 'Target', 'Value']]

```

```python
# Get the original "Paired" color palette from Seaborn
original_palette = sns.color_palette()

# Define a function to generate shades with varying lightness
def generate_lightness_range(color):
    shades = []
    for i in range(1, 5):
        lightness = 0.1 + 0.2 * i  # Adjust lightness from 0.1 to 0.9 in increments of 0.2
        light_color = sns.set_hls_values(color, l=lightness)  # Set lightness component
        shades.append(light_color)
    return shades[::-1]

# Create a custom color palette with four shades of each color in a range of lightness
custom_palette = []
for color in original_palette:
    custom_palette.extend(generate_lightness_range(color))


# Convert the color tuples to a list of color strings in Hex format
custom_palette_hex = [plt.cm.colors.to_hex(c) for c in custom_palette]
sns.palplot(custom_palette_hex)

```
